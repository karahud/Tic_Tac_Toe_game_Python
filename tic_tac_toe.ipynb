```python
# Below is the python code to create a tic-tac-toe game

# Function to create an empty board
def create_grid():
    print("Here is the playboard:")
    board = [[" ", " ", " "],
             [" ", " ", " "],
             [" ", " ", " "]]
    return board

# Function to print the board in a formatted way
def print_board(board):
    for row in board:
        print("  |  ".join(row))  # Prints each row with separators
        print("---+-----+---")  # Prints row separator
    return board

# Function to assign symbols to players
def symbol():
    print('Select "X" or "O" as your symbol.')

    while True:  # Ensures valid input
        player1 = input('Player 1, select your symbol: ').upper()
        if player1 == "X":
            player2 = "O"
            break
        elif player1 == "O":
            player2 = "X"
            break
        else:
            print('Invalid input! Please select either "X" or "O".')

    print(f'Player 1, your symbol is "{player1}".')
    print(f'Player 2, your symbol is "{player2}".')
    return player1, player2

# Function to take user input and update the board
def start(board, player1, player2, count):
    player = player1 if count % 2 == 0 else player2  # Determines current player
    print(f"{player}, it is your turn.")

    while True:
        try:
            row = int(input('Enter the row (0 = upper, 1 = middle, or 2 = lower): '))
            col = int(input('Enter the column (0 = left, 1 = middle, or 2 = right): '))

            if 0 <= row <= 2 and 0 <= col <= 2:  # Ensures valid range
                if board[row][col] == " ":  # Checks if space is available
                    board[row][col] = player
                    break
                else:
                    print("That square is already taken. Choose another.")
            else:
                print('Invalid input! Please enter a number between 0 and 2.')
        except ValueError:
            print("Please enter a valid number.")

# Function to check for a winner
def winner(board):
    # Check rows
    for row in board:
        if row[0] == row[1] == row[2] != " ":
            return row[0]

    # Check columns
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] != " ":
            return board[0][col]

    # Check diagonals
    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]

    return None  # No winner yet

# Function to check for a tie
def tie(board):
    for row in board:
        if ' ' in row:
            return False  # If any empty space remains, game is not tied
    return True  # No empty spaces, game is tied

# Main game loop
board = create_grid()
print_board(board)
player1, player2 = symbol()
count = 0  # Track number of moves

while True:
    print_board(board)
    start(board, player1, player2, count)

    winner_symbol = winner(board)  # Store winner result to avoid redundant calls
    if winner_symbol:
        print_board(board)
        print(f"Player {winner_symbol} wins!")
        print(f'Congratulations {winner_symbol}, you have won the game!')
        break
    elif tie(board):
        print_board(board)
        print('Game Over!')
        print("It's a tie!")
        print('Better luck next time.')
        print('Thanks for playing!')
        break
    count += 1
```
